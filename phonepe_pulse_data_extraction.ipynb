# **Aggregated transaction by states**

```python
import pandas as pd

aggtran_state_dict = {'State':[],
                      'Year':[],
                      'Quater':[],
                      'Transaction_type':[],
                      'Transaction_count':[],
                      'Transaction_amount':[]}
for i in aggtran_state_dir:
  state_path = path+i+'/'
  state_year = os.listdir(state_path)
  for j in state_year:
    year_path = state_path+j+'/'
    year_dir = os.listdir(year_path)
    for k in year_dir:
      jsonfile = year_path+k
      a = open(jsonfile,'r')
      Data = json.load(a)
      for z in Data['data']['transactionData']:
        name = z['name']
        count = z['paymentInstruments'][0]['count']
        amount = z['paymentInstruments'][0]['amount']
        aggtran_state_dict['Transaction_type'].append(name)
        aggtran_state_dict['Transaction_count'].append(count)
        aggtran_state_dict['Transaction_amount'].append(amount)
        aggtran_state_dict['State'].append(i)
        aggtran_state_dict['Year'].append(j)
        aggtran_state_dict['Quater'].append('Q'+k.strip('.json'))

df_agg_tran_state = pd.DataFrame(aggtran_state_dict)


```

```python
df_agg_tran_state
```

```python
df_agg_tran_state.to_csv('agg_tran_state.csv',index=False)
```

# **Aggregated user by states**

```python
path = "/content/pulse/data/aggregated/user/country/india/state/"
aggtran_user_dir = os.listdir(path)

aggtran_user_dict = {'State':[],
                      'Year':[],
                      'Quater':[],
                      'Brand':[],
                      'Brand_count':[],
                      'Brand_percentage':[]}
for i in aggtran_user_dir:
  state_path = path+i+'/'
  state_year = os.listdir(state_path)
  for j in state_year:
    year_path = state_path+j+'/'
    year_dir = os.listdir(year_path)
    for k in year_dir:
      jsonfile = year_path+k
      a = open(jsonfile,'r')
      Data = json.load(a)
      try:
        for x in Data['data']['usersByDevice']:
          brand = x['brand']
          count = x['count']
          percentage = x['percentage']

          aggtran_user_dict['Brand'].append(brand)
          aggtran_user_dict['Brand_count'].append(count)
          aggtran_user_dict['Brand_percentage'].append(percentage)
          aggtran_user_dict['State'].append(i)
          aggtran_user_dict['Year'].append(j)
          aggtran_user_dict['Quater'].append('Q'+k.strip('.json'))
      except TypeError:
        pass




df_agg_user = pd.DataFrame(aggtran_user_dict)
df_agg_user

```

```python
df_agg_user.to_csv('agg_tran_user.csv',index=False)
```

# **Map Transaction by States**

```python
path='/content/pulse/data/map/transaction/hover/country/india/state/'
aggtran_user_dir = os.listdir(path)

map_tran_state = {'State':[],
                      'Year':[],
                      'Quater':[],
                      'District':[],
                      'Transaction_count':[],
                      'Transaction_amount':[]}
for i in aggtran_user_dir:
  state_path = path+i+'/'
  state_year = os.listdir(state_path)
  for j in state_year:
    year_path = state_path+j+'/'
    year_dir = os.listdir(year_path)
    for k in year_dir:
      jsonfile = year_path+k
      a = open(jsonfile,'r')
      Data = json.load(a)
      for y in Data['data']['hoverDataList']:
        name = y['name']
        count = y['metric'][0]['count']
        amount = y['metric'][0]['amount']
        map_tran_state['District'].append(name)
        map_tran_state['Transaction_count'].append(count)
        map_tran_state['Transaction_amount'].append(amount)
        map_tran_state['State'].append(i)
        map_tran_state['Year'].append(j)
        map_tran_state['Quater'].append('Q'+k.strip('.json'))

df_map_tran_state = pd.DataFrame(map_tran_state)
df_map_tran_state

```

```python
df_map_tran_state.to_csv('map_tran_state.csv',index=False)
```

# **Map User by States**

```python
path='/content/pulse/data/map/user/hover/country/india/state/'
aggtran_user_dir = os.listdir(path)

map_user_disct = {'State':[],
                      'Year':[],
                      'Quater':[],
                      'District':[],
                      'RegisteredUsers':[],
                      'AppOpens':[]}
for i in aggtran_user_dir:
  state_path = path+i+'/'
  state_year = os.listdir(state_path)
  for j in state_year:
    year_path = state_path+j+'/'
    year_dir = os.listdir(year_path)
    for k in year_dir:
      jsonfile = year_path+k
      a = open(jsonfile,'r')
      Data = json.load(a)
      for w in Data['data']['hoverData']:
        name = w
        registeruser = Data['data']['hoverData'][w]['registeredUsers']
        app_opening = Data['data']["hoverData"][w]["appOpens"]
        map_user_disct['District'].append(name)
        map_user_disct['RegisteredUsers'].append(registeruser)
        map_user_disct['AppOpens'].append(app_opening)
        map_user_disct['State'].append(i)
        map_user_disct['Year'].append(j)
        map_user_disct['Quater'].append('Q'+k.strip('.json'))

df_map_user = pd.DataFrame(map_user_disct)
```
